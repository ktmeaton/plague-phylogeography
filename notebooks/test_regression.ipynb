## Import Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

## DataFrame Setup

```python
df = pd.DataFrame({
                  "dist2root" : [1,5,2,3,4,5],
                  "date": ["2014-01-01", "2015-01-02","2015-03-03", "2016-06-06", "2017-07-09", "2018-10-12"]
                 })

df['date_ordinal'] = pd.to_datetime(df['date']).apply(lambda date: date.toordinal())
```

## Create Basic Scatterplot

```python
plt.figure()
points = plt.scatter(x=df['date'], y=df['dist2root'])
plt.show()
```

## Add Regression Line

```python
ax = sns.regplot(x="date_ordinal",
                       y="dist2root",
                       data=df,
                       ci=90)
```

```python
ax = sns.regplot(x="date_ordinal",
                       y="dist2root",
                       data=df,
                       scatter=True,
                       fit_reg=True,
                       ci=90,
                       order=1)
```
